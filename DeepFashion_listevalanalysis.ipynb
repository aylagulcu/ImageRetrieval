```python

class reader():

    def __init__(self,image1,image2,item_id,evalution_status):
        self.image1 = image1
        self.image2 = image2
        self.item_id = item_id
        self.evalution_status = evalution_status

    def __repr__(self) -> str:
        return self.image1 + " " + self.image2 + " " + self.item_id + " " + self.evalution_status

class Dataset():
        

    def __init__(self,root):
        self.count_dict ={"test":0 , "train" : 0 ,"validation" : 0}
        self.datadict =  {"test": [],"train" : [], "validation" : []}
        self.posnegdict = {"test":{"pos":0,"neg":0},"train":{"pos":0,"neg":0},"validation":{"pos":0,"neg":0}}
        delimiter = " "
        with open(root, 'r') as file:

            # Skip the first two lines
            next(file)
            next(file)

            # Read the rest of the lines one by one
            for line in file:
                # Do something with the line
                lineedited = " ".join(line.split())
                item = reader(*lineedited.split())
                if item.evalution_status == "test":
                    self.count_dict["test"] = self.count_dict.get("test",0) + 1
                    self.datadict["test"].append(item)
                elif item.evalution_status == "train":
                    self.count_dict["train"] = self.count_dict.get("train",0) + 1
                    self.datadict["train"].append(item)
                elif item.evalution_status == "val":
                    self.count_dict["validation"] = self.count_dict.get("validation",0) + 1
                    self.datadict["validation"].append(item)


        for mode in ["test","train","validation"]:
            for data in self.datadict[mode]:
                if data.image1.split("/")[3] == data.image2.split("/")[3]  and data.image1.split("/")[1] == data.image2.split("/")[1] and data.image1.split("/")[2] == data.image2.split("/")[2]  : 
                    self.posnegdict[mode]["pos"] = self.posnegdict[mode].get("pos",0) + 1
                else : 
                    self.posnegdict[mode]["neg"] = self.posnegdict[mode].get("neg",0) + 1
```
